### *Cleaning Data*

Berdasarkan Insight yang telah didapatkan dari Assessing Data, maka dapat dilakukan cleaning data untuk masing masing masalah yang ditemukan pada setiap dataset


#### **Membersihkan Duplicates Dataset Geolocation**

````python
print("Jumlah duplikasi: ", geolocation_df.duplicated().sum())
````

Metode yang dipilih untuk mengatasi duplikat data tersebut adalah drop_duplicates

```
# This is formatted as code
```



````python
geolocation_df.drop_duplicates(inplace=True)
print("Jumlah duplikasi: ", geolocation_df.duplicated().sum())
````

####**Mengubah Tipe Data pada Dataset Order Item**


Kolom "shipping_limit_date" yang berbentuk data object akan diubah menjadi datetime dengan syntax dan library dari Pandas: *pd.to_datetime*

````python
order_items_df["shipping_limit_date"] = pd.to_datetime(order_items_df["shipping_limit_date"])
print(order_items_df.info())
````

####**Mengatasi Missing Value pada Dataset Order Review**

##### Missing Value Review Comment Title

Dilakukan peninjauan mengenai kolom *review_comment_title* yang tidak bernilai NaN atau tidak memiliki isi

````python
order_reviews_df[order_reviews_df.review_comment_title.isna()]
````

Dilakukan peninjauan mengenai baris kolom *review_comment_title* yang memiliki nilai atau tidak bernilai NaN

````python
order_reviews_df.review_comment_title.value_counts()
````

Dilakukan pengisian baris kolom *review_comment_title* dengan isi "no title" untuk menangani proses missing value

````python
order_reviews_df.review_comment_title.fillna(value="no title", inplace=True)
print(order_reviews_df.isna().sum())
````

##### Missing Value Review Comment Message

Dilakukan peninjauan mengenai kolom *review_comment_message* yang tidak bernilai NaN atau tidak memiliki isi

````python
order_reviews_df[order_reviews_df.review_comment_message.isna()]
````

Dilakukan peninjauan mengenai baris kolom *review_comment_message* yang memiliki nilai atau tidak bernilai NaN

````python
order_reviews_df.review_comment_message.value_counts()
````

Dilakukan pengisian baris kolom *review_comment_message* dengan isi "no message" untuk menangani proses missing value

````python
order_reviews_df.review_comment_message.fillna(value="no message", inplace=True)
print(order_reviews_df.isna().sum())
````

#####Mengubah Tipe Data (Object menjadi Datetime)

````python
datetime_columns = ["review_creation_date", "review_answer_timestamp"]

for column in datetime_columns:
  order_reviews_df[column] = pd.to_datetime(order_reviews_df[column])

print(order_reviews_df.info())
````

## **Visualization & Explanatory Analysis**

### Negara bagian manakah yang jumlah customer paling banyak dan paling sedikit dalam lingkup e-commerce ini

````python
state_df = customer_df.groupby("customer_state").customer_id.nunique().sort_values(ascending=False).reset_index()
state_df
````

````python
sns.barplot(x="customer_id", y="customer_state", data=state_df.head(27))
plt.show()
````

### Metode pembayaran apa yang sering dipakai dalam bertransaksi pada e-commerce

````python
methods_df = order_payments_df.groupby("payment_type").order_id.nunique().sort_values(ascending=False).reset_index()
methods_df
````

````python
sns.barplot(x="order_id", y="payment_type", data=methods_df.head(27))
plt.show()
````

### Produk jenis apakah yang memiliki tingkat penjualan tertinggi dan terendah

````python
highest_product_df = english_item_df.groupby("product_category_name_english").product_id.nunique().sort_values(ascending=False).reset_index()
highest_product_df
````

````python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="product_id", y="product_category_name_english", data=highest_product_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("5 Produk Penjualan Tertinggi", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)

sns.barplot(x="product_id", y="product_category_name_english", data=highest_product_df.sort_values(by="product_id", ascending=True).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("5 Produk Penjualan Terendah", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)

plt.suptitle("Data Penjualan Produk Tertinggi dan Terendah", fontsize=20)
plt.show()
````

### Bagaimana tingkat kepuasan customer terhadap e-commerce
